Create an input folder **"input_images"**, and save the images to be analyzed inside this folder. The folder **output_images** will be auto created. If there exist this folder with files, the folder will be emptied before the new analysis.

```python
import os
import shutil
from ultralytics import YOLO
model_yolo = YOLO("yolov8n.pt")

input_dir = "input_images"
output_dir = "output_images"
if os.path.isdir(output_dir):
    for item in os.listdir(output_dir):
        item_path = os.path.join(output_dir, item)
        if os.path.isdir(item_path):
            shutil.rmtree(item_path)
        else:
            os.remove(item_path)

os.makedirs(output_dir, exist_ok=True)
os.makedirs(input_dir, exist_ok=True)
```

```python
import numpy as np
import cv2
req_conf = np.float64(0.3)
for image_name in os.listdir(input_dir):
  img_path = os.path.join(input_dir, image_name)
  img = cv2.imread(img_path)
  results = model_yolo(img_path)
  result = results[0]
  boxes = result.boxes
  names = result.names
  for box in boxes:
    cls_id = int(box.cls)
    print(f"Cls ID is {cls_id}")
    conf = np.float64(box.conf)
    print(f"Conf is {conf}")
    if (cls_id == 0) and (conf>=req_conf):
      x1,y1,x2,y2 = map(int, box.xyxy[0])
      cv2.rectangle(img, (x1, y1), (x2, y2), color=(0,255,0), thickness=2)
      label = f"{names[cls_id]} {conf: .2f}"
      cv2.putText(img, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)
      output_path = os.path.join(output_dir, image_name)
      cv2.imwrite(output_path, img)

```
